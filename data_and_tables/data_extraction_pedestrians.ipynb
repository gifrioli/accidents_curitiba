# Data Extraction
My data was obtained through Brazil's Access to Information Law. The documents were submitted in PDF format, so I decided to use the Natural PDF library, created by Professor Jonathan Soma, to extract the numbers. Library documentation is available at: https://jsoma.github.io/natural-pdf/

```python
#Extracting only data on accidents involving pedestrians in 2020
```

```python
from natural_pdf import PDF

pdf_path = 'atropelamentos_curitiba/ATROPELAMENTO_PEDESTRES_2020.pdf'
pdf = PDF(pdf_path)
```

```python
all_dfs = []

for page_number in range(len(pdf.pages)):
    page = pdf.pages[page_number]
    
    table = page.extract_table(method="pdfplumber")
    
    if table:
        
        df = table.to_df()
        all_dfs.append(df)
    else:
        print(f"No tables found on page {page_number + 1}")

import pandas as pd

df_completo = pd.concat(all_dfs, ignore_index=True)

print(df_completo)
```

```python
csv_path = 'acidentes_pedestres_2020.csv'

#Save in CSV
df_completo.to_csv(csv_path, index=False, encoding='utf-8-sig')
```

# I saved the tables into one and added the year.

```python
years = [2020, 2021, 2022, 2023, 2024, 2025]
dfs = []

for year in years:
    archive = f'acidentes_pedestres_{year}.csv'
    df = pd.read_csv(archive)
    df['Ano'] = year
    dfs.append(df)

df_combine = pd.concat(dfs, ignore_index=True)
df_combine.to_csv('acidentes_pedestres_todos_anos.csv', index=False, encoding='utf-8-sig')
```
